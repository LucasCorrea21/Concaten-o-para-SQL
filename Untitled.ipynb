```python
import pandas as pd

data_df = pd.read_csv('AirBnB.csv', sep=",")

data_df = data_df[['Data','Tipo','Noites','Quantia','Taxa do Anfitrião','Taxa de Limpeza']]

filtrandotipo = data_df[data_df['Tipo'] == 'Reserva']
filtrandotipo = filtrandotipo.sort_values('Noites', ascending = False)
#display(filtrandotipo)


filtrandotipo.plot('Noites', 'Quantia',  kind= 'bar', figsize=(16,5), title ='Índice de Aluguel')
```
